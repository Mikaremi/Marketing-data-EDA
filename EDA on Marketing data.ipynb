# Exploratory data analysis


## average age of company's customers

```python
avg_age = marketing_df['Age'].mean()
avg_age
```

the average age of customers in that company is 51

## Age vs customer ratio

```python
age_df = marketing_df.groupby('Age')[['Age']].count()
age_df
```

```python
age_graph = plt.figure(figsize=(12,6))
sns.barplot(x=age_df.index,y=age_df.Age)
plt.ylabel('Count')
plt.xticks(rotation = 90);
```

## campaing aquisition

```python
# how many customers are accepted in campaign 1 as per the total customers
cmpper1 = marketing_df.AcceptedCmp1.sum()/len(marketing_df)*100
cmpper2 = marketing_df.AcceptedCmp2.sum()/len(marketing_df)*100
cmpper3 = marketing_df.AcceptedCmp3.sum()/len(marketing_df)*100
cmpper4 = marketing_df.AcceptedCmp4.sum()/len(marketing_df)*100
cmpper5 = marketing_df.AcceptedCmp5.sum()/len(marketing_df)*100

print('''
Customer Offer Aquisition on campaign 1 is {:.2f}%
Customer Offer Aquisition on campaign 2 is {:.2f}%
Customer Offer Aquisition on campaign 3 is {:.2f}%
Customer Offer Aquisition on campaign 4 is {:.2f}%
Customer Offer Aquisition on campaign 5 is {:.2f}%
'''.format(cmpper1,cmpper2, cmpper3,cmpper4,cmpper5))
```

## average complaint ratio

```python
complaint_ratio = len(marketing_df['Complain'] == 1)/len(marketing_df)
complaint_ratio
```

```python
print('complaint ratio of company is {:.2f}%'.format(complaint_ratio*100))
```

## Average purchases on varient outlets and categories

```python
web_purchases = marketing_df['NumWebPurchases'].mean()
store_purchases = marketing_df['NumStorePurchases'].mean()
catalog_purchases = marketing_df['NumCatalogPurchases'].mean()
discount_purchases = marketing_df['NumDealsPurchases'].mean()

print('''
Average Purchase on website is: {}
Average Purchase on store is: {}
Average Purchase on catalog is: {}
Average Purcahse with discount is: {}
'''.format(web_purchases, store_purchases, catalog_purchases, discount_purchases))
```

## Avearage recency of purchase

```python
avg_rec = marketing_df['Recency'].mean()
print('Average purchase day of company is {:.2f}'.format(avg_rec))
```

## customer website visit ratio

```python
visit_web = len(marketing_df[marketing_df['NumWebVisitsMonth']>0])/len(marketing_df)
print('customer website visit percentage  {:.2f}'.format(visit_web*100))  
```

# Question: Do they need wheelchair accessibility in the store building?

## count senior citizen ration from store customers. Only (senior citizen age is more than 55)

```python
#narrow it down to the store visitors only
store_customers = marketing_df[marketing_df['NumDealsPurchases']>0]
```

```python
# calculate from store customers that are senior citizens
senior_citizens = store_customers[store_customers['Age']>55]
```

```python
#calculate an average of senior citizens customers
avg_scc = len(senior_citizens)/len(marketing_df)*100

print('average % of senior citizen customrs is: {:.2f}'.format(avg_scc))
```

There is a good number of senior citizens that visit the store, thus it is imparative for there to be a wheelchair accesibility in the company's store
